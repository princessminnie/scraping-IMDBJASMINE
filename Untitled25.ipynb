```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://www.imdb.com/chart/toptv?pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=470df400-70d9-4f35-bb05-8646a1195842&pf_rd_r=V9GS13YW8K5AHQ9T47KA&pf_rd_s=right-4&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_ql_6"
```

```python
page = requests.get(url)
page
```

```python
page.content
```

```python
soup = BeautifulSoup(page.content, "html.parser")
print(soup.prettify())
```

```python
scraped_movies = soup.find_all('td', class_='titleColumn')
scraped_movies
```

```python
movies = []
for movie in scraped_movies:
    movie = movie.get_text().replace('\n', "")
    movie = movie.strip(" ")
    movies.append(movie)
movies
```

```python
scraped_ratings = soup.find_all('td', class_='ratingColumn imdbRating')
scraped_ratings
```

```python
ratings = []
for rating in scraped_ratings:
    rating = rating.get_text().replace('\n', "")
    ratings.append(rating)
ratings
```

```python
data = pd.DataFrame()
data['Movie Names'] = movies
data['Ratings'] = ratings
data.head()
```

```python
data.to_csv('IMDB Top Rate TV Shows.csv', index=False)
```
